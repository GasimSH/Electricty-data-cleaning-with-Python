```python
#Her bir serfiyyatini qiymetine vurub cemleyirik. Sonda illik serfiyyati aliriq
kwh_month_sum=0
for i in range(12):
    kwhx=df_month.loc[i,"KWh_x"]
    kwhy=df_month.loc[i,"KWh_y"]
    kwh_month=kwhx*kwhy
    kwh_month_sum+=kwh_month
kwh_month_sum=round(kwh_month_sum,2)
print(kwh_month_sum)
```

```python
# Eger saatliq paketi gotursek netice ne olacaq;

df_hoursly=df.groupby("hours")["KWh"].sum()
df_hoursly=pd.merge(df_hoursly, hoursly, on=["hours"], how='left')
print(df_hoursly)
```

```python
# Yene serfiyyati qiymete vurub cemleyirik, saatliq pakete gore illik xercimizi tapiriq

kwh_hours_sum=0
#print(df_hoursly)
for i in range(24):
    #print(i)
    kwhx=df_hoursly.loc[i,"KWh_x"]
    kwhy=df_hoursly.loc[i,"KWh_y"]
    df_hours=kwhx*kwhy
    #print(kwhx, kwhy, df_hours)
    kwh_hours_sum+=df_hours
print(round(kwh_hours_sum,2))
```
